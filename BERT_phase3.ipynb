```python
from bertopic import BERTopic
from umap import UMAP
from sklearn.feature_extraction.text import CountVectorizer
from bertopic.vectorizers import ClassTfidfTransformer
from bertopic.representation import KeyBERTInspired
from bertopic.representation import MaximalMarginalRelevance
```

```python
# to get the same result every time
umap_model = UMAP(
    #n_neighbors=3,
    #n_components=5,
    #min_dist=0.0,
    #metric='cosine',
    random_state=42)
```

```python
vectorizer_model = CountVectorizer(
    min_df=0.08, # ignore terms that appear in less than 8% of the documents
    max_df = 0.15, # ignore terms that appear in more than 15% of the documents. Hopefully we'll get rid of (old, elder, senior, etc.)
    ngram_range=(1, 2), # allowing to get values like "British Columbia" instead of "British", and "Columbia".
    stop_words="english" # removing english stop words
    )
```

```python
ctfidf_model = ClassTfidfTransformer(
    bm25_weighting=True, # At smaller datasets, this variant can be more robust to stop words that appear in your data
    reduce_frequent_words=True # To further reduce these frequent words
    )
```

```python
# The main representation of a topic
main_representation = KeyBERTInspired()

# Additional ways of representing a topic
add_representation = [MaximalMarginalRelevance(diversity=.1)]

# Add all models together to be run in a single `fit`
representation_model = {
   "Main": main_representation,
   "Aspect1":  add_representation
}
```

```python
seed_topic_list = [
    ['candidate', 'election', 'government', 'law', 'minister', 'president', 'senator', 'state', 'vote'],
    ['coronavirus', 'covid19', 'dose', 'hospital', 'immunity', 'nurse', 'outbreak', 'treatment', 'vaccine', 'wellness'],
    ['celebrate', 'ceremony', 'christmas', 'economy', 'hate', 'immigration', 'indian', 'labour', 'poverty', 'prejudice', 'protests', 'retired', 'transgender', 'veteran', 'wage', 'wheelchair'],
    ['athlete', 'club', 'hockey', 'jersey', 'player', 'team']
    ]
```

# BERTopic (1000 texts)

```python
model = BERTopic(
    umap_model = umap_model,
    vectorizer_model=vectorizer_model,
    ctfidf_model=ctfidf_model,
    representation_model=representation_model,
    seed_topic_list=seed_topic_list, # to guide the topics towards the seed_topic_list
    nr_topics = 10
    )
```

```python
topics, probs = model.fit_transform(texts)
```

# Metrics

##Coherence score
https://github.com/MaartenGr/BERTopic/issues/469

```python
from gensim import corpora
from gensim.corpora import Dictionary
from gensim.models.coherencemodel import CoherenceModel
```

```python
# Extract vectorizer and analyzer from BERTopic
vectorizer = model.vectorizer_model
analyzer = vectorizer.build_analyzer()
```

```python
tokens = [analyzer(text) for text in texts]
dictionary = corpora.Dictionary(tokens)
corpus = [dictionary.doc2bow(token) for token in tokens]
topic_words = [[words for words, _ in model.get_topic(topic)] for topic in range(len(set(topics))-1)]
```

```python
coherence_model = CoherenceModel(
    texts=tokens,
    dictionary=dictionary,
    corpus=corpus,
    topics = topic_words,
    coherence='c_v')
```

```python
coherence_score = coherence_model.get_coherence()
print("Coherence Score:", coherence_score)
```

The coherence score is a measure of how coherent the topics generated by the BERTopic model are. A coherence score typically ranges from 0 to 1, where higher values indicate more coherent and meaningful topics.

The coherence score of 0.4945 suggests that the topics generated by the BERTopic model are moderately coherent. A score around 0.5 is generally considered reasonable, but it can vary depending on the field and dataset.

Since this is unsupervised learning, we always need human judgement to evaluate the output of our models.

```python
coherence_per_topic = coherence_model.get_coherence_per_topic()
coherence_per_topic
```

As explained before, the higher the score, the more coherent it is. Topics 1, 2, and 5 have relatively high coherence scores (0.6419, 0.6019, and 0.7860, respectively), indicating that they are well-formed and coherent topics. We could focus on these well-formed topics; consider them well defined topics within our collection of tweets. In the other hand, those topics with low values might need revisions and/or further tuning of the model.

Topic modeling is an exploratory process, and achieving the most coherent and relevant topics may require fine-tuning and experimentation. It's essential to achieve a balance between coherence and topic diversity while ensuring the topics are relevant. Furthermore, we must always manually inspect the tweets associated to each topic to ensure the topic is relevant.


##Coherence score for different models

```python
# create a dataframe to store the results
coherence_results = pd.DataFrame(columns=["Run", "NumTopics", "Count-1", "Coherence", "CoherencePerTopic"])
```

```python
coherence_results
```

```python
# Run a few times and store the results
for i in range(8,16):
  try:
    # BERTopic model
    model = BERTopic(umap_model = umap_model,
                    vectorizer_model=vectorizer_model,
                    ctfidf_model=ctfidf_model,
                    representation_model=representation_model,
                    seed_topic_list=seed_topic_list, # to guide the topics towards the seed_topic_list
                    nr_topics = i # to variate the number of topics in each run
      )

    # Fit BERTopic and get the topics
    topics, probs = model.fit_transform(texts)

    # Get the number of topics and count of tweets in topic -1
    num_topics = len(model.get_topic_freq())
    count_minus_1 = topics.count(-1)

    # Get coherence per model and per model.topic
    ### Extract vectorizer and analyzer from BERTopic
    vectorizer = model.vectorizer_model
    analyzer = vectorizer.build_analyzer()

    ### Extract features for Topic Coherence evaluation
    tokens = [analyzer(text) for text in texts]
    dictionary = corpora.Dictionary(tokens)
    corpus = [dictionary.doc2bow(token) for token in tokens]
    topic_words = [[words for words, _ in model.get_topic(topic)] for topic in range(len(set(topics))-1)]

    ### Evaluate Coherence
    coherence_model = CoherenceModel(texts=tokens,
                                    dictionary=dictionary,
                                    corpus=corpus,
                                    topics = topic_words,
                                    coherence='c_v')
    coherence_score = coherence_model.get_coherence()
    coherence_per_topic = coherence_model.get_coherence_per_topic()

    # Append the results to the DataFrame
    coherence_results = coherence_results.append({
        "Run": i ,
        "NumTopics": num_topics,
        "Count-1": count_minus_1,
        "Coherence": coherence_score,
        "CoherencePerTopic": coherence_per_topic
        }, ignore_index=True)

  except ValueError as e:
        print(f"Skipping iteration due to error: {e}")
        continue
```

```python
coherence_results
```

```python
for coherence_score in coherence_results['CoherencePerTopic']:
    formatted_scores = ["{:.2f}".format(score) for score in coherence_score]
    print(formatted_scores)
```

The above data frame shows that the models with fewer topics tends to have more coherence.

As mentioned before, topic modeling is an unsupervised learning technique so the results can be ambiguous, difficult to interpret, and they are sensitive to the data fed.
