```python
def clean_tweet(tweet):
    # Remove Twitter handles
    tweet = re.sub(r'@[A-Za-z0-9_]+', '', tweet)

    # Handle encoding issues (if any)
    tweet = tweet.encode('utf-8', 'ignore').decode('utf-8')

    # Remove unnecessary characters
    tweet = re.sub(r'[^A-Za-z0-9\s]', '', tweet)

    # Correct spelling and abbreviations
    tweet = str(TextBlob(tweet).correct())

    # Tokenize the text (using NLTK)
    #tweet_tokens = word_tokenize(tweet)

    # Apply additional preprocessing steps
    #tweet_tokens = [word.lower() for word in tweet_tokens if word not in stop_words]

    return tweet
```

```python
import nltk
from transformers import BertTokenizer
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('wordnet')
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    # Tokenize the text using BERT tokenizer
    tokens = tokenizer.tokenize(text)

    # Lemmatize each token
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Convert the lemmatized tokens back to a string
    processed_text = ' '.join(lemmatized_tokens)

    return processed_text
```

```python
from bertopic import BERTopic
from umap import UMAP
from sklearn.feature_extraction.text import CountVectorizer
from bertopic.vectorizers import ClassTfidfTransformer
from bertopic.representation import KeyBERTInspired
from bertopic.representation import MaximalMarginalRelevance
```

```python
ctfidf_model = ClassTfidfTransformer(
    bm25_weighting=True, # At smaller datasets, this variant can be more robust to stop words that appear in your data
    reduce_frequent_words=True # To further reduce these frequent words
    )
```

```python
# The main representation of a topic
main_representation = KeyBERTInspired()

# Additional ways of representing a topic
add_representation = [MaximalMarginalRelevance(diversity=.5)]

# Add all models together to be run in a single `fit`
representation_model = {
   "Main": main_representation,
   "Aspect1":  add_representation
}
```

Defining a function to get the tweets from a certain topic, from a given model's result.

```python
# Before running this function, you first need to define what the document_info is. For example:
# "document_info = model.get_document_info(texts)"
def get_tweets(document_info, topic_nr):
    topic = document_info[document_info['Topic'] == topic_nr]
    for _, row in topic.iterrows():
        tweet = row['Document']
        probability = row['Probability']
        print(f'Tweet: {tweet}\nProbability: {probability}\n')
```

####Guided
https://maartengr.github.io/BERTopic/getting_started/guided/guided.html

This method guides BERT towards certain keywords (seeds) that we know are in the documents and are representative of a specific topic. In our case the topics are politics, health, socio-econ, and entertainment.

We pass the seeds with words related to each of these topics and BERT will try to assign the tweets to one of the seeds.

The output of this BERT variant is several topics, where at least 4 correspond to each of the seeds we planted.

####Semi-Supervised
https://maartengr.github.io/BERTopic/getting_started/semisupervised/semisupervised.html

Semi-supervised modeling allows us to steer the dimensionality reduction of the embeddings into a space that closely follows any labels you might already have.

######preparing labels

Using the results from the model_guided output, we will label each of the tweets according to the topic I think they fall under.

The following will be our labels:
* 1: politics
* 2: health
* 3: socioeconomic_generic
* 4: socioeconomic_mobility
* 5: socioeconomic_first.nations
* 6: socioeconomic_ethnicity
* 7: entertainment_sports
* 8: entertainment_movies

The tweets from each topic from the *guided* result will be assigned the following labels:
* Topic 0: label 1 - politics (if prob = 1)
* Topic 1: N/A
* Topic 2: label 2 - health
* Topic 3: label 3 - Socioeconomic_generic (if prob = 1)
* Topic 4: label 7 - entertainment_sports (if prob > 0.5)
* Topic 5: label 4 - socioeconomic_mobility
* Topic 6: label 5 - socioeconomic_first.nations
* Topic 7: label 8 - entertainment_movies
* Topic 8: label 6 - socioeconomic_ethnicity

```python
def generate_label(row):
    topic = row['Topic']
    probability = row['Probability']

    if topic == 0 and probability == 1:
        return 1
    elif topic == 2:
        return 2
    elif topic == 3 and probability == 1:
        return 3
    elif topic == 4 and probability > 0.5:
        return 7
    elif topic == 5:
        return 4
    elif topic == 6:
        return 5
    elif topic == 7:
        return 8
    elif topic == 8:
        return 6
    else:
        #(for topics -1 and 1)
        return -1
```

```python
document_info_guided['Label'] = document_info_guided.apply(generate_label, axis=1)
```

####Supervised
https://maartengr.github.io/BERTopic/getting_started/supervised/supervised.html

In this case we need defined labels. Instead of trying to predict the underlying topics in our model, we pass the predifined topics (deifned by labels) and we BERT will try to learn the relationship between those topics and the input texts.

In other words, we will do a classification task with the help of Logistic Regression.

To do so, we need to skip over the dimensionality reduction step and replace the clustering step with a classification algorithm.

#Conclusions

* Trying to clean the tweets as ChatGPT does is very hard. Even with help from this AI, it is hard to achieve the same result.
* Lemmatizing did help to get rid of redundant words in the topic representations. However, the sentences were butchered, which I believe would cause syntax to be lost. The documentation actually recommends not to do any pre-processing.
* Although setting a random state and other parameters in the umap_model section helps to maintain reproducibility, modifying another paramter like min_df, ngram_range or something else in any of the other modules can cause new results. I think all parameters must always be fixed to get the exact same result. When I was running Guided and Semi, I got an error asking me to increase the max_df parameter. That's why I started initializing the umap_model and vectorizer_model modules at the beginning of each new model.
* Having a lot of repeated or very similar tweets can make a whole topic for them.
* Guided is a good technique push BERT to cluster the texts towards some topics that you think may exist.
* Semi-supervised is the tool to push BERT to cluser texts towards some topics that are already labeled. New topics may also be generated.
* Supervised is a classification technique. This does not help to label unlabeled data. By contrast, it is a powerful tool to compare the output of 2 different models.
